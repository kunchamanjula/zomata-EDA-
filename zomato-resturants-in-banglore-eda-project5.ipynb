# 1.what are the budget friendly resturants for users with good taste in banglore ?

```python
budget_friendly=zomato[["restaurant name","avg cost (two people)","rate (out of 5)"]].sort_values(by=["avg cost (two people)","rate (out of 5)"],ascending=[True,False]).head()
budget_friendly
```

```python
sns.barplot(x=budget_friendly["avg cost (two people)"],y=budget_friendly["restaurant name"],hue=budget_friendly["rate (out of 5)"],data=zomato)
plt.title('Budget Friendly Restaurants in Banglore  w.r.t  Rating and cost')
plt.show()
```

# 2.Recommend resturant type for new enterpretenur?

**TOP 10 Recommended resturant types for new resturant owner**

```python
top_resturant_type=zomato['restaurant type'].value_counts().head(10) #getting top 10 retutrant types by using sort values
top_resturant_type
```

```python
explode=[0.1,0,0,0,0,0.3,0.3,0.3,0.3,0.3]
plt.pie(x=top_resturant_type,labels=top_resturant_type.index,autopct="%1.1f%%",explode=explode) #plotting pie chart to get the visualization of top 10 resturant types
plt.title('Top 10 Resturant types in Banglore')

plt.show()
```

**BOTTOM 7 NOT  Recommended resturant types for new resturant owner**

```python
bottom_resturant_type=zomato['restaurant type'].value_counts().tail(7)#getting BOTTOM 7 retutrant types by using sort values
bottom_resturant_type
```

```python

plt.pie(x=bottom_resturant_type,labels=bottom_resturant_type.index,autopct="%1.1f%%")#plotting pie chart to get the visualization BOTTOM 7 resturant types
plt.title('Bottom 7 Resturant types in Banglore')

plt.show()
```

# 3. plot the barchart to  withrespect to index and differeint varaibles in dataset?

```python
zomato_variables = ["rate (out of 5)","avg cost (two people)"]

fig, axs = plt.subplots(2, figsize=(12, 10))

for index, ax in enumerate(fig.axes):
    sns.barplot(y=zomato[zomato_variables[index]].value_counts(), 
                x=zomato[zomato_variables[index]].value_counts().index, ax=ax)
    ax.set_ylabel("Count")
    ax.set_xlabel(zomato_variables[index])

plt.show()
```

# 5.how many resturants has online booking options in banglore?

```python
online=zomato["online_order"].value_counts()
online
```

```python

plt.pie(online,labels=online.index,autopct='%1.1f%%')
cen_cir=plt.Circle((0,0),0.70,fc='white')
fig=plt.gcf()
fig.gca().add_artist(cen_cir)
plt.axis('equal')
plt.title('Percentage of restaurants that offer online order')
plt.show()
```

* There are equal number of resturants which has online bookings and also not  online booking resturants

# 6.provide the top 10 list of resturants in banglore based on their ratings and reviews given by zomato users? 

```python
top_rated=zomato[['restaurant name','num of ratings','rate (out of 5)']].sort_values(by=['num of ratings','rate (out of 5)'],ascending=[False,False]).head(10)
top_rated
```

```python
sns.barplot(x=top_rated['num of ratings'],y=top_rated['restaurant name'],hue=top_rated['rate (out of 5)'])
plt.title('Top restaurants based on reviws and ratings')
plt.show()
```

# 7.provide the top 10 fastfood and beverages restaurants in Bangalore?

```python
fastfood_bev=zomato.groupby('cuisines type').get_group('Fast Food, Beverages')[['restaurant name','rate (out of 5)','num of ratings']].sort_values(by=['rate (out of 5)','num of ratings'],ascending=[False,False]).head(10)
fastfood_bev
```

```python
sns.barplot(x=fastfood_bev['num of ratings'],y=fastfood_bev['restaurant name'],hue=fastfood_bev['rate (out of 5)'])
plt.title('Top 10 fastfood and bevaerages Restaurant In bangalore')
plt.show()
```
